```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,VotingClassifier
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,f1_score

X_train, X_test, y_train, y_test = train_test_split(data[columns_final], data['Outcome'], test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier(random_state=42)
lr = LogisticRegression(random_state=42)
svc = SVC(C=10,kernel='linear')
xg = xgb.XGBClassifier(learning_rate = 0.1,max_depth= 2,colsample_bytree = 0.4,min_child_weight = 2, n_estimators = 50,subsample = 0.6)
```

```python
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
svc.fit(X_train, y_train)
xg.fit(X_train, y_train)

# Make predictions
pred_rf = rf.predict(X_test)
pred_lr = lr.predict(X_test)
pred_svc = svc.predict(X_test)
pred_xgb = xg.predict(X_test)

# Evaluate the models
accuracy_rf = accuracy_score(y_test, pred_rf)
accuracy_lr = accuracy_score(y_test, pred_lr)
accuracy_svc = accuracy_score(y_test, pred_svc)
accuracy_xgb = accuracy_score(y_test, pred_xgb)



# Print the accuracies
print(f"Random Forest Accuracy: {accuracy_rf}")
print(f"Logistic Regression Accuracy: {accuracy_lr}")
print(f"SVM Accuracy: {accuracy_svc}")
print(f"XGBoost Accuracy: {accuracy_xgb}")
```

```python
f1_rf = f1_score(y_test, pred_rf)
f1_lr = f1_score(y_test, pred_lr)
f1_svc = f1_score(y_test, pred_svc)
f1_xgb = f1_score(y_test, pred_xgb)

# Print the F1-Scores
print(f"Random Forest F1-Score: {f1_rf}")
print(f"Logistic Regression F1-Score: {f1_lr}")
print(f"SVM F1-Score: {f1_svc}")
print(f"XGBoost F1-Score: {f1_xgb}")
```

```python
#Tuning hyperparameters process begins

param_rf = {
    'n_estimators':[50,100,200,300],
    'max_depth':[None,8,12,20],
    'min_samples_split': [2, 5, 10]
}

param_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}

param_svc = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf']
}

param_xgb = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6, 9]
}
```

```python
grid_rf = GridSearchCV(rf, param_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_lr = GridSearchCV(lr, param_lr, cv=5, scoring='accuracy', n_jobs=-1)
grid_svc = GridSearchCV(svc, param_svc, cv=5, scoring='accuracy', n_jobs=-1)
grid_xgb = GridSearchCV(xg, param_xgb, cv=5, scoring='accuracy', n_jobs=-1)

grid_rf.fit(X_train, y_train)
grid_lr.fit(X_train, y_train)
grid_svc.fit(X_train, y_train)
grid_xgb.fit(X_train, y_train)

# Get the best estimators
best_rf = grid_rf.best_estimator_
best_lr = grid_lr.best_estimator_
best_svc = grid_svc.best_estimator_
best_xgb = grid_xgb.best_estimator_
```

```python
#RandomForestClassifier(max_depth=8, min_samples_split=5, random_state=42)
final_rf = RandomForestClassifier(random_state=42,max_depth=8,min_samples_split=5)

#LogisticRegression(C=1, penalty='l1', random_state=42, solver='liblinear')
final_lr = LogisticRegression(random_state=42,C=1,penalty='l1',solver='liblinear')

final_svc = SVC(C=10,kernel='linear',gamma='auto')


final_xg = xgb.XGBClassifier(learning_rate = 0.1,colsample_bytree = 0.4,min_child_weight = 2, n_estimators = 50,subsample = 0.6,max_depth = 9)
```

```python
final_clf = VotingClassifier(estimators=[
    ('rf', final_rf), 
    ('lr', final_lr), 
    ('xgb', final_xg),
    ('svc',final_svc)],
    voting='hard')
```

```python
final_clf.fit(X_train,y_train)
```

```python
final_voting = final_clf.predict(X_test)

accuracy = accuracy_score(y_test,final_voting)
```

```python
print(accuracy)
```

```python
from sklearn.ensemble import StackingClassifier

#Base models
#C prevents misclassification

estimators_intermediate = [
    ('lr', final_lr),
    ('svc', final_svc),
]
```

```python
#Last model
last_model = final_rf

stacking = StackingClassifier(estimators=estimators_intermediate,final_estimator=last_model,cv=5)
```

```python
stacking.fit(X_train,y_train)
```

```python
pred_stacking = stacking.predict(X_test)

# Evaluate the model
accuracy_stacking = accuracy_score(y_test, pred_stacking)

# Print the accuracy
print(f"Stacking Classifier Accuracy: {accuracy_stacking}")
```
